```python
import pandas as pd
import io
import requests
import seaborn as sns
import timeit
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/carlosgilrodrigo/08MBID---Visualizacion-de-Datos/main/Database_Limpio.csv"
s=requests.get(url).content
premier=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
premier
```

```python
premier.info()
```

```python
premier.rename(columns={'HC':'Córners_Equipo_Local'},inplace=True)
premier.rename(columns={'AC':'Córners_Equipo_Visitante'},inplace=True)
```

```python
import numpy as np
sns.histplot(premier.Córners_Equipo_Local.dropna(), kde=True, binwidth=0.5, linewidth=1.5, color='#F47528', bins=5, binrange=(0, 20))
plt.xticks(ticks=np.arange(0.25, 20, 1), labels=np.arange(0, 20, 1))

plt.title("Córners a favor del equipo local por partido", fontsize=16, fontweight='bold')
plt.xlabel("Córners del equipo local", fontsize=12, fontweight='bold')
plt.ylabel("Número de partidos", fontsize=12, fontweight='bold')
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.show()
```

```python
import numpy as np
sns.histplot(premier.Córners_Equipo_Visitante.dropna(), kde=True, binwidth=0.5, linewidth=1.5, color='#4F12A8', bins=5, binrange=(0, 20))
plt.xticks(ticks=np.arange(0.25, 20, 1), labels=np.arange(0, 20, 1))

plt.title("Córners a favor del equipo visitante por partido", fontsize=16, fontweight='bold')
plt.xlabel("Córners del equipo visitante", fontsize=12, fontweight='bold')
plt.ylabel("Número de partidos", fontsize=12, fontweight='bold')
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.show()
```

```python
premier.rename(columns={'FTHG':'Goles_Local'},inplace=True)
premier.rename(columns={'FTAG':'Goles_Visitante'},inplace=True)
```

```python
premier
```

```python
premier["Resultado"] = premier["Goles_Local"].apply(str) + "-" + premier["Goles_Visitante"].apply(str)
```

```python
premier.head()
```

```python
resultados_comunes = premier["Resultado"].value_counts().nlargest(10)
```

```python
resultados_comunes.head()
```

```python
plt.figure(figsize=(14, 7))
graf = sns.barplot(x=resultados_comunes.index, y=resultados_comunes.values, palette="rocket")
total=len(premier)

for i in graf.patches:
    percentage = '{:.1f}%'.format(100 * i.get_height() / total)
    x = i.get_x() + i.get_width() / 2 - 0.17
    y = i.get_y() + i.get_height() + 5
    graf.annotate(percentage, (x, y), size=14)


plt.title("Resultados más comunes en la Premier League", fontsize=18, fontweight='bold')
plt.xlabel("Resultado", fontsize=14, fontweight='bold')
plt.ylabel("Cantidad de partidos", fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.show()
```
